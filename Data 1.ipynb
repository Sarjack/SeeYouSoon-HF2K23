```python
import pandas as pd
import numpy as np
import snscrape.modules.twitter as sntwitter
import datetime
```

```python
def fetcher(comp, start, end):
  query = f'${comp} until:{end} since:{start}'
  lis = []
  limit = 50
  for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(lis) == limit:
      break
    else:
      try:
        if tweet.lang == 'en' and tweet.cashtags == [comp]: #and #tweet.viewCount > 10:
          lis.append([tweet.rawContent])
      except:
        pass
  data = pd.DataFrame(lis,columns= ['Content'])
  return data
```

```python
data = fetcher('AVGO','2018-01-01','2018-01-02')
data
```
